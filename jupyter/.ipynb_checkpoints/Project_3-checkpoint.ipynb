```python
#Import all dependencies
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
csv_file_alcohol_consumption = "HappinessAlcoholConsumption.csv"
df = pd.read_csv(csv_file_alcohol_consumption) 
df.head()
```

```python
df['Total Alcohol Consumption']= df.iloc[:, 6:9].sum(axis=1)
df.head()
```

```python
#New total database
df_totals = df[['Country', 'Region','HappinessScore','HDI','GDP_PerCapita','Total Alcohol Consumption']].copy()
df_totals.head()
```

```python
#New Region dataset
df_region =df_totals.groupby(['Region']).sum()
df_region = df_region.reset_index()
df_region = df_region.sort_values(by='HappinessScore', ascending=False)
df_region = df_region.reset_index()
df_region = df_region.drop(['index'],axis= 1)
df_region = df_region.reset_index()
df_region = df_region.rename(columns={'index':'id_Region'})
df_region['id_Region'] = df_region['id_Region']+1

df_region
```

```python
#New Top 10 countries dataset
df_countries = df_totals.sort_values(by='HappinessScore', ascending=False)
df_countries = df_countries.head(20)
df_countries
```

```python
#Merge to add region ids to Country dataset
df_region_2= df_region[['id_Region','Region']]
df_countries_2 = df_countries.merge(df_region_2, left_on='Region', right_on='Region', how='left')
df_countries_2 = df_countries_2.reset_index()
df_countries_2 = df_countries_2.rename(columns={'index':'id_Country'})
df_countries_2['id_Country'] = df_countries_2['id_Country']+1
df_countries_2
```
